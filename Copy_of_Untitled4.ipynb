Project Description: Create a simple to-do list manager using Python. The program will allow users to add tasks, mark tasks as completed, list tasks, and remove tasks.


Requirements:

## Basic User Interface:
Upon starting the program, display a welcome message and a list of available actions: "1. Add Task", "2. Mark Task as Completed", "3. List Tasks", "4. Remove Task", and "5. Quit".
#Task List:
Implement a way to store tasks. You can use a list or any other appropriate data structure.
#Add Task:
When the user selects the "Add Task" option, prompt them to enter the task description.
Add the task to the task list.
#Mark Task as Completed:
Allow the user to mark a task as completed by entering its index in the task list.
If the task is already marked as completed, display an appropriate message.
#List Tasks:
Display a numbered list of tasks along with their completion status.
#Remove Task:
Allow the user to remove a task by entering its index in the task list.
#Quit:
When the user selects the "Quit" option, display a goodbye message and terminate the program.

```python
tasks = []
```

```python
def home():
  """
  a function that will enter the actions needed.

  input:
  the actions ; to add,list,remove,mark as complete and quit

  return:
  home
  """
  print("\n Welcome to Todo List\n1. Add task\n2. Mark as completed\n3. List Task\n4. Remove Tasks\n5. Quit ")

def add_task():
  task_name = input("Enter Task Description: ")
  tasks.append({"description": task_name,"completed":False})
  print(f"Your task,{task_name} was added successfully")
#funtion for adding tasks
def mark_complete():
  try: #to havve a nested loop
    task_index = int(input("Enter the task index: "))
    if 0 <= task_index < len(tasks):#toensure that the index is within the lenghth of the list
      if not tasks[task_index]["completed"]:#to check whether the task has been marked complete
        tasks[task_index]["completed"] = True #if all the condtions are the its changed to marked
        print("Task has been marked complete.")
      else:
        print("The task is already completed!")
    else:
      print("Oops, index entered out of range!")

  except ValueError:
    print("Kindly enter valid task index!")

def list_tasks():
  for index, task in enumerate(tasks):#to give the list indices.
    status = "Completed" if task["completed"] else "Not Completed" #this is too list down the list whether completed or not
    print(f"{index},{task['description']},{status}")

def remove_task():
  list_tasks()
  try: #nested loop
    task_index = int(input("Enter the task index: "))
    if 0 <= task_index < len(tasks): #to ensure it is the range
      del tasks[task_index]
    else:
      print("Oops, index entered out of range!")

  except ValueError:
    print("Kindly enter valid task index!")

def run_task_app():
  while True: #using a while loop to ensure that input is run across all our functioins
    home()
    selection = input("Select an option from the menu: ")
    if int(selection) == 1:
      add_task()
    elif int(selection) == 2:
      mark_complete()
    elif int(selection) == 3:
      list_tasks()
    elif int(selection) == 4:
      remove_task()
    elif int(selection) == 5:
      print("Goodbye!")
      break
    else:
      print("Oopsie!Invalid input.")
```

```python

run_task_app()
```
